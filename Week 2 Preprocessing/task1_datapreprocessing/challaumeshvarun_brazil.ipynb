```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import interpolate
from sklearn.preprocessing import *
```

```python
df=pd.read_csv('/content/north.csv')
numof_zeros=(df==0).sum()
numof_na=df.isna().sum()
print(numof_zeros)
```

```python
# replace 0 and -9999 with NaN
df.replace(-9999, pd.NA, inplace=True)
df.replace(0, pd.NA, inplace=True)
print(df)
#filling NaN with mean
df[df.columns[3:20]]=df[df.columns[3:20]].fillna(df[df.columns[3:20]].mean())
print(df)
```

Standardisation

```python
scaler = StandardScaler()

# Standardize columns 4 to 20
stdised_data = scaler.fit_transform(df[df.columns[3:20]])
# Creating a DataFrame from the standardized data
df_standardized = pd.DataFrame(stdised_data, columns=df[df.columns[3:20]].columns)
print(df_standardized)
```

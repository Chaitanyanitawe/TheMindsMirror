```python
import numpy as np
import pandas as pd
import os
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
sc = StandardScaler()
```

```python
folder_path = '/content/drive/My Drive/data'
file_list = os.listdir(folder_path)
xlsx_files = [f for f in file_list if f.endswith('.xlsx')]
for xlsx_file in xlsx_files:
    input_filepath = os.path.join(folder_path, xlsx_file)
    output_filepath = os.path.join(folder_path, f"{os.path.splitext(xlsx_file)[0]}.csv")
    df = pd.read_excel(input_filepath)
    df.to_csv(output_filepath, index=False)
```

```python
csv_files = [f for f in file_list if f.endswith('.csv')]
for csv_file in csv_files:
    input_file_path = os.path.join(folder_path, csv_file)
    output_file_path = os.path.join(folder_path, f"{os.path.splitext(csv_file)[0]}_normalized.csv")
    df = pd.read_csv(input_file_path)
    df.isnull().sum()
    imputer= SimpleImputer(missing_values =np.nan, strategy='mean')
    new_columns = df.columns[0:]
    df[new_columns] = sc.fit_transform(df[new_columns])
    df.to_csv(output_file_path, index=False)
    print(df.head())
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mpt
```

```python
file_path='/content/north.csv'
```

```python
data=pd.read_csv(file_path)
data['Data']=pd.to_datetime(data['Data'])

#getting one years data
data_set=data.loc[data['Data'].dt.year == 2000]

data_set.head()
```

```python
#checking null values/
data_set.isnull().sum()
```

```python
#replacing -9999.0 with null values (is clearly wrong data)
data_set.replace(-9999,pd.NA)
```

```python
#replacing null value with mean value of the particular column
columns_fill=data_set.columns[3:20]
mean=data_set[columns_fill].mean()
data_set.loc[:, columns_fill] = data_set[columns_fill].fillna(mean)
data_set.head()

```

```python
data_set.isnull().sum()
```

```python
#dropping region,state,station_code,latitude,longitude,height since they are the same

data_set.drop(["region","state",'station','station_code','latitude','longitude','height'],axis=1)
```

```python
#removing outliers using z score
from scipy import stats


def remove_outliers_zscore(df, threshold=4):

    numeric_columns = df.select_dtypes(include=['number'])
    z_scores = stats.zscore(numeric_columns)

    # z-score measures how many standard deviations a data point is from the mean.if greater than 4 then data point is an outlier
    outliers = (abs(z_scores) > threshold).any(axis=1)

    df_no_outliers = df[~outliers]

    return df_no_outliers

data_set_new = remove_outliers_zscore(data_set)
```

```python
#standardizing data: data in each column has a mean of 0 and standard deviation of 1
#standardising columns[3:20]

columns_to_standardize=data_set_new.columns[3:20]


from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_standardized = scaler.fit_transform(data_set_new[columns_to_standardize])

df_standardized = pd.DataFrame(df_standardized, columns=[f'{col}_standardized' for col in columns_to_standardize])

#Adding standardized values to data frame
df = pd.concat([data_set_new, df_standardized], axis=1)

# Droping original values from data frame
df.drop(columns=columns_to_standardize, inplace=True)


df.head()


```

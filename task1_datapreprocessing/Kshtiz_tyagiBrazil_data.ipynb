```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import os
import time
from scipy.fft import fft, ifft
from scipy import interpolate
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv('north.csv')
```

```python
print(len(df))
df.head()
```

```python
zero_counts = (df == 0).sum()
print(zero_counts)
nan_count = df.isna().sum()
print(nan_count)
```

Filling the zero balues in the data using interpolate

```python
# Fill zeros with NaN
df.replace(0, pd.NA, inplace=True)
df.replace(-9999, pd.NA, inplace=True)
# Fill NaN values with linear interpolation
df.interpolate(method='linear', axis=0, inplace=True)
zero_counts = (df == 0).sum()
print(zero_counts)
nan_count = df.isna().sum()
print(nan_count)
```

Filling the null values with the median of the column for rows 3 on wards

```python
df[df.columns[3:20]] = df[df.columns[3:20]].fillna(df[df.columns[3:20]].median())
```

```python
df.head()
```

Standardizing the data

```python
scaler = StandardScaler()
standardized_data = scaler.fit_transform(df[df.columns[3:20]])
standardized_df = pd.DataFrame(standardized_data, columns=df[df.columns[3:20]].columns)
standardized_df.head()
```

Normalizing the data

```python
scaler = MinMaxScaler()

# Fit the scaler to the data and transform the DataFrame
normalized_data = scaler.fit_transform(df[df.columns[3:20]])

# Create a new DataFrame with the normalized data
normalized_df= pd.DataFrame(normalized_data, columns=df[df.columns[3:20]].columns)
normalized_df.head(100)
```

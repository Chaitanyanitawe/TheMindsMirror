### Applying linear interpolation

```python
from scipy import interpolate
# Function to resample an array to the target length
def resample_array(array, target_length):
    # Create an array of indices for the input array
    input_indices = np.linspace(0, len(array)-1, len(array))

    # Create an array of indices for the resampled array
    resampled_indices =np.linspace(0,len(array)-1,target_length)

    # Create a linear interpolation function based on the input array
    interpolator =interpolate.interp1d(input_indices,array)

    # Use the interpolator to create the resampled array
    resampled_array =interpolator(resampled_indices)

    return resampled_array
```

```python
count=0
for df in filtered_data:
    channel1=np.array(df[['Output1']]).flatten()
    channel2=np.array(df[['Output2']]).flatten()
    channel3=np.array(df[['Output3']]).flatten()
    channel4=np.array(df[['Output4']]).flatten()
    channel5=np.array(df[['Image']]).flatten()
    channel1=resample_array(channel1,mean_length)
    channel2=resample_array(channel2,mean_length)
    channel3=resample_array(channel3,mean_length)
    channel4=resample_array(channel4,mean_length)
    channel5=resample_array(channel5,mean_length)
    data={
        "Output1":channel1,
        "Output2":channel2,
        "Output3":channel3,
        "Output4":channel4,
        "Image":channel5
    }
    filtered_data[count]=pd.DataFrame(data)
    count+=1
print(filtered_data)    
```

### Using SMOTE Upsampling for Data Augmentation

```python
from imblearn.over_sampling import SMOTE

combined_df = pd.concat(filtered_data)

x=combined_df[['Output1','Output2','Output3','Output4']]
y=combined_df[['Image']]

smote = SMOTE(sampling_strategy=0.8, random_state=42)
X_resampled, y_resampled = smote.fit_resample(x, y)

resampled_df = pd.concat([X_resampled, y_resampled], axis=1)

def split_dataframe(df, chunk_size):
    return [df[i:i + chunk_size] for i in range(0, df.shape[0], chunk_size)]

filtered_data=split_dataframe(resampled_df, 1266)

filtered_data
```

```python
x=[]
y=[]
for df in filtered_data:
    arr1=df[['Output1','Output2','Output3','Output4']].values.tolist()
    arr2=df[['Image']].values.tolist()
    x.append(arr1)
for i in range(len(filtered_data)):
    y.append(filtered_data[i].iloc[0]['Image'])
x=np.array(x)
y=np.array(y)
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train_val,x_test,y_train_val,y_test=train_test_split(x,y,test_size=0.1)
x_train,x_val,y_train,y_val=train_test_split(x_train_val,y_train_val,test_size=0.1)
print(x_train.shape,x_val.shape,x_test.shape)
print(y_train.shape,y_val.shape,y_test.shape)
```

### Building a 1D CNN

```python
import tensorflow as tf
from tensorflow.keras import models, layers
model=models.Sequential([
    layers.Conv1D(32,30,input_shape=(x_train.shape[1],4)),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling1D(pool_size=2,strides=2),
    layers.Conv1D(64,30),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling1D(pool_size=2,strides=2),
    layers.Conv1D(128,30),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling1D(pool_size=2,strides=2),
    layers.Conv1D(256,30),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling1D(pool_size=2,strides=2),
    layers.Conv1D(512,30),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling1D(pool_size=2,strides=2),
    layers.GlobalAveragePooling1D(),
    layers.Dense(90,activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(50,activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(10,activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(2)
])
model.summary()
```

```python
loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)

lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate=1e-2,
    decay_steps=60,
    decay_rate=0.48,
    staircase=True)
optimizer=tf.keras.optimizers.Adam(learning_rate=lr_schedule)
```

```python
metrics=['accuracy']
model.compile(loss=loss,optimizer=optimizer,metrics=metrics)
callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=7,restore_best_weights=True)
```

```python
history=model.fit(x_train,y_train,epochs=50,batch_size=20,shuffle=True,verbose=2,validation_data=(x_val,y_val),callbacks=callback)
```

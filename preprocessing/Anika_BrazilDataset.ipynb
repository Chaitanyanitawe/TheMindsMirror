```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv('/content/drive/My Drive/north.csv')
print(df.dtypes)
df['Data']=pd.to_datetime(df['Data'])
df['year']=df['Data'].dt.year
print(df)
```

```python
for yr in df['year'].unique():
  data=df[df['year']==yr]
  print(yr,data['station_code'].nunique())
data=df[df['year']==2007]
print(data.isnull().sum())
```

```python
print(len(data))
random_select=np.random.choice(data['station_code'].unique(),10,replace=False)
fin_data=data[data['station_code'].isin(random_select)]
print(len(fin_data))
print(fin_data)
```

```python
fin_data.isnull().sum()
```

```python
fin_data=fin_data.replace(-9999,np.nan)
fin_data.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer= SimpleImputer(missing_values =np.nan, strategy='mean')
```

```python
fin_data.head()
```

```python

selected_columns = fin_data.columns[3:20]
imputer.fit(fin_data[selected_columns])
fin_data[selected_columns] = imputer.transform(fin_data[selected_columns])
```

```python
fin_data.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
columns_to_scale = fin_data.columns[3:20]
fin_data[columns_to_scale]=sc.fit_transform(fin_data[columns_to_scale])
fin_data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
columns_to_scale = fin_data.columns[3:20]
fin_data[columns_to_scale]=sc.fit_transform(fin_data[columns_to_scale])
fin_data.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[1,2,20,21,22,23])],remainder='passthrough')
fin_data=np.array(ct.fit_transform(fin_data))
```
